```python
import numpy as np
from ast import literal_eval
import pandas as pd
from sklearn.metrics import adjusted_rand_score
from classifiers import RandomForestForCancer
```

```python
with open("coad_cluster_labels.txt",'r') as file:
    cluster_labels = literal_eval(file.readline())
```

```python
len(cluster_labels)
```

```python
rf_can = RandomForestForCancer(get_tr_mats=False,csvpath="COAD_snv.tsv")
```

```python
rf_can.create_gold_coad()
rf_can.gold_df
```

```python
rf_can.load_data(csvpath="COAD_snv.tsv")
```

```python
rf_can.sample_gene_matrix.shape
```

```python
X,y = rf_can.create_training_matrix()
```

```python
len(y)
```

```python
true_labels = y.tolist()
pred_labels = cluster_labels
len(true_labels),len(pred_labels)
```

```python
na_indices = [i for i in range(len(true_labels)) if true_labels[i]==-1]

for i in range(len(true_labels)-1,-1,-1):
    if i in na_indices:
        del true_labels[i]
        del pred_labels[i]
```

```python
ari = adjusted_rand_score(true_labels, pred_labels)
ari
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
nmi = normalized_mutual_info_score(true_labels, pred_labels)
nmi
```
