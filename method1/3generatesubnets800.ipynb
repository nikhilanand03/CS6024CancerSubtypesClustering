```python
from subnetwork_generator import SubnetworkGenerator
from ast import literal_eval
```

```python
with open("top800.txt",'r') as file:
    list_of_genes = file.readline()
    print(literal_eval(list_of_genes))
```

```python
arr = literal_eval(list_of_genes)
arr[0]
```

```python
sg = SubnetworkGenerator(list_of_genes=arr)
```

```python
nodes,adj = sg.get_graph(sg.get_interactions(sg.list_of_genes))
```

```python
clusts = sg.get_clusters(adj)
```

```python
sg.draw_graph(nodes,sg.get_clustering_dict(nodes,clusts),sg.get_interactions(sg.list_of_genes))
```

```python
clust_dict = sg.get_clustering_dict(nodes,clusts)
```

```python
clust_dict
```

```python
clusters = {}
for key in clust_dict.keys():
    cluster = clust_dict[key][0]
    try:
        clusters[cluster].append(key)
    except:
        clusters[cluster] = [key]
```

```python
with open("cluster_dict.txt",'w') as file:
    file.write(str(clusters))
```
