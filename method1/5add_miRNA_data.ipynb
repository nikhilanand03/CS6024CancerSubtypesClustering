```python
from ast import literal_eval
import numpy as np
```

```python
data_matrix = []
with open("sample_x_subnetwork.txt",'r') as file:
    lines = file.readlines()
    for line in lines:
        data_matrix.append(literal_eval(line))
data_matrix = np.array(data_matrix)
```

```python
from miRNAProcessor import miRNAProcessor
```

```python
mrp = miRNAProcessor("COAD")
```

```python
transf_matrix = mrp.load_miRNA()
```

```python
with open("samples.txt",'r') as file:
    samples = literal_eval(file.readline())
```

```python
len(samples)
```

```python
combined_multiomic = mrp.get_combined_multiomic(data_matrix,transf_matrix,samples)
```

```python
with open("coad_subnetwork_mutations_miRNA.txt",'w') as file:
    file.write(str(combined_multiomic.tolist()))
```
