```python
import numpy as np
import pandas as pd
from ast import literal_eval
import seaborn as sns
```

```python
with open("brca_differential_mutations_miRNA.txt",'r') as file:
    brca_diff_muts = np.array(literal_eval(file.readline()))

brca_diff_muts.shape
```

```python
with open("coad_differential_mutations_miRNA.txt",'r') as file:
    coad_diff_muts = np.array(literal_eval(file.readline()))

coad_diff_muts.shape
```

```python
brca_diff_muts, coad_diff_muts
```

```python
sns.heatmap(brca_diff_muts)
```

```python
with open("brca_cluster_labels.txt",'r') as file:
    label = literal_eval(file.readline())
    arr = list(enumerate(label))
    tuples = sorted(arr, key=lambda x: x[1])
    reordered_indices = [x[0] for x in tuples]
    reordered_matrix = brca_diff_muts[reordered_indices]
sns.heatmap(reordered_matrix)
```
